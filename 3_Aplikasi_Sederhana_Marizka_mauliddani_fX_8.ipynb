#Aplikasi Menghitung Luas Persegi Panjang

```python
panjang = int(input("Masukkan panjang dari persegi panjang (cm): "))
lebar = int(input("Masukkan lebar dari persegi panjang (cm): "))

luas = panjang * lebar

print(f"\nLuas dari persegi panjangnya adalah {luas} cm^2.")
```

#Aplikasi Menghitung Volume Bangun[Tabung]

```python
jari = int(input("Masukan jari-jari dari tabung (cm) : ") )
tinggi = int(input("Masukan tinggi dari tabung (cm) : ") )

volume = 22/7 * jari * jari * tinggi

print(f"\n volume dari tabung yaitu {volume} cm^3")
```

#Aplikasi Menghitung Luas Bangun Datar [Tabung]

```python
jari = int(input("Masukkan jari-jari dari tabung (cm): "))
tinggi = int(input("Masukkan tinggi dari tabung (cm): "))


luas = 2 * 22/7 * ( jari+tinggi )


print(f"\n luas dari tabung yaiu {luas} cm^2")
```
