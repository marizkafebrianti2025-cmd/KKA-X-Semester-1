#Cara Penulisan Operasi Aritmatika pada Python

```python
#Cara 1. Hitung langsung di print

print (24*3)
```

```python
#Cara 2: Simpan angka ke variabel, lalu di tampilkan

total_jumlah_botol = 24 * 3

print(total_jumlah_botol)
```

```python
# Cara 3: Simpan angka ke variabel, lalu kalikan langsung di print

banyak_botol_per_kardus = 24
jumlah_kardus = 3

print(banyak_botol_per_kardus * jumlah_kardus)
```

```python
#Cara 4: Simpan angka ke variabel, hasilnya juga disimpan ke variabel baru, lalu tampilkan

banyak_botol_per_kardus = 24
jumlah_kardus = 3
total_jumlah_botol = banyak_botol_per_kardus * jumlah_kardus

print(total_jumlah_botol)
```
